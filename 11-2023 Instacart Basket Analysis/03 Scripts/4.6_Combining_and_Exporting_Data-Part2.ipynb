A suitable way to merge the two dataframes df_prods and df_ords_prods_combined is by the column product_id, found in both dataframes.

```python
# Drop unnamed columns in df_prods (not necessary) 
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# Drop unnamed columns and _merge column in df_ords_prods_comb (not necessary) 
df_ords_prods_comb = df_ords_prods_comb.drop(columns = ['Unnamed: 0', '_merge'])
```

```python
# Product_id matches in both dataframes - use this as the key to merge
df_ords_prods_merged = df_ords_prods_comb.merge(df_prods, on = 'product_id', indicator = True)
```
