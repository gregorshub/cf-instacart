## Step 5: Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# Checking for missing values and summing them

customer.isnull().sum()
```

There are 11259 missing values in the first_name column.

```python
# Checking for duplicates

customer[customer.duplicated()]
```

There are no duplicates.

```python
# Checking to see if there are any mixed types
for col in customer.columns.tolist():
    weird = (customer[[col]].applymap(type) != customer[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customer[weird]) > 0:
        print (col)
```

The column first_name has mixed types which is probably due to the missing values.

```python
# Looking at the missing values
customer.loc[customer['first_name'].isna()]
```

```python
# Addressing missing values by imputation
customer['first_name'].fillna('Unknown', inplace = True)
```

```python
# Rechecking missing values
customer.isnull().sum() 
```

```python
# Rechecking mixed types
for col in customer.columns.tolist():
    weird = (customer[[col]].applymap(type) != customer[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customer[weird]) > 0:
        print (col)
```

After replacing the missing values in the first_name column, there are no mixed types and to missing values anymore.
