## Step 3: You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the â€œorder_hour_of_dayâ€ column.

    In a markdown cell beneath your histogram, describe what the histogram demonstrates.


```python
# Creating a histogram on 'order_hour_of_day' column using 24 bins for 24 hours in a day

hist2 = ords_prods_customer_merge['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram illustrates that in the hours between 9 am and 4 pm most of the orders are placed whereas in the hours between 1 am and 6 am the least orders are conducted. Furthermore, there is a steady decrease of the orders placed after 4 pm.

```python
# Exporting histogram
hist2.figure.savefig(os.path.join(path, '04 Analysis',  'Visualizations', 'hist_order_hour_of_day.png'))
```
